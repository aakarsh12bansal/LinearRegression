```python
def gradient(x,y,w,b):
    '''dj_dw is an array of m elements with dj_dw[i] 0<=i<m
       dj_db is also an array of m '''
    f=np.dot(x,w)+b
    k=len(y)
    J=f-y
    m=x.shape[1]
    #print(f)
    dj_dw=np.zeros(m)
    '''
    now, to compute gradient:
    
    J=sum((fi-y)**2)/2m, where f=w1x_i1+w2x_i2+w3x_i3 +b 
    therefore,
     dj_dw1= (1/m)*sum(x1*(fi-y)) '''
    dj_db=np.sum(J)/k
    #print(dj_db)
    #print(x[:,0])
    #for i in range(m):
    dj_dw= np.dot((f-y),x)

    dj_dw/=k
   # print(dj_dw)
    return dj_dw,dj_db


gradient(x,y,w,b)
```

```python
def gradient_descent(x,y,w,b,alpha):
    wt=w
    bt=b

    for i in range(10000):
        dj_dw,dj_db=gradient(x,y,wt,bt)
        wt= wt - alpha*dj_dw

        bt= bt - alpha*dj_db
        
        if i%1000==0:
            print(f"cost: {cost(x,y,wt,bt)}")
    
    return wt,bt
```
